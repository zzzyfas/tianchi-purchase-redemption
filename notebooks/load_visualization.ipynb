```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 读取并处理日期
data = pd.read_csv('../data/raw/Purchase Redemption Data/user_balance_table.csv', engine='python')
data = data[['report_date', 'total_purchase_amt', 'total_redeem_amt']]
data['report_date'] = pd.to_datetime(data['report_date'].astype(str), format='%Y%m%d')

# 2. 关键：按日期分组求和，得到每日汇总数据
data_daily = data.groupby('report_date')[['total_purchase_amt', 'total_redeem_amt']].sum().reset_index()

# 3. 按时间排序并重置索引（确保时间连续）
data_daily = data_daily.sort_values('report_date').reset_index(drop=True)
```

```python
# 绘图（使用汇总后的数据）
plt.figure(figsize=(25, 4))
plt.plot(data_daily['report_date'], data_daily['total_purchase_amt'], label='total_purchase_amt')
plt.plot(data_daily['report_date'], data_daily['total_redeem_amt'], label='total_redeem_amt')
plt.grid(True, linestyle="-", color="green", linewidth="0.5")
plt.legend()
plt.title('purchase and redeem of every month')

plt.gca().spines["top"].set_alpha(0.0)
plt.gca().spines["bottom"].set_alpha(0.3)
plt.gca().spines["right"].set_alpha(0.0)
plt.gca().spines["left"].set_alpha(0.3)

plt.show()
```

```python
import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler

# 固定随机种子
np.random.seed(2019)


class RNNModel(object):

    def __init__(
        self,
        look_back=40,
        epochs_purchase=150,
        epochs_redeem=230,
        batch_size=16,
        verbose=2,
        patience=50,
        store_result=True
    ):

        self.look_back = look_back
        self.epochs_purchase = epochs_purchase
        self.epochs_redeem = epochs_redeem
        self.batch_size = batch_size
        self.verbose = verbose
        self.patience = patience
        self.store_result = store_result

        # ==========================================
        # 读取原始数据
        # ==========================================
        data = pd.read_csv(
            '../data/raw/Purchase Redemption Data/user_balance_table.csv',
            engine='python'
        )

        # 只保留需要列
        data = data[
            [
                'report_date',
                'total_purchase_amt',
                'total_redeem_amt'
            ]
        ]

        # 日期转换
        data['report_date'] = pd.to_datetime(
            data['report_date'].astype(str),
            format='%Y%m%d'
        )

        # ==========================================
        # 按日期聚合
        # ==========================================
        data_daily = data.groupby('report_date')[
            [
                'total_purchase_amt',
                'total_redeem_amt'
            ]
        ].sum().reset_index()

        # 日期排序
        data_daily = data_daily.sort_values(
            'report_date'
        ).reset_index(drop=True)

        # 可视化检查
        print(data_daily.head())
        print(f'\n总天数: {len(data_daily)}')

        # ==========================================
        # 保存为 date_label.csv（可选）
        # ==========================================
        data_daily.to_csv(
            'date_label.csv',
            index=False
        )

        print('\ndate_label.csv 已生成')

        # ==========================================
        # 分离 purchase / redeem
        # ==========================================
        self.purchase = data_daily[
            ['total_purchase_amt']
        ]

        self.redeem = data_daily[
            ['total_redeem_amt']
        ]

        # 保存日期用于绘图
        self.date_daily = data_daily


    # ==================================================
    # 数据预处理
    # ==================================================
    def access_data(self, data_frame):

        data_set = data_frame.values
        data_set = data_set.astype('float32')

        # 归一化
        scaler = MinMaxScaler(feature_range=(0, 1))
        data_set = scaler.fit_transform(data_set)

        # 构造时序数据
        train_x, train_y, test = self.create_data_set(data_set)

        # reshape
        train_x = np.reshape(
            train_x,
            (train_x.shape[0], 1, train_x.shape[1])
        )

        return train_x, train_y, test, scaler


    # ==================================================
    # 构造数据集
    # 前40天 -> 预测未来30天
    # ==================================================
    def create_data_set(self, data_set):

        data_x = []
        data_y = []

        for i in range(
            len(data_set) - self.look_back - 30
        ):

            x = data_set[
                i:(i + self.look_back),
                0
            ]

            y = data_set[
                i + self.look_back:
                i + self.look_back + 30,
                0
            ]

            data_x.append(x)
            data_y.append(list(y))

        data_x = np.array(data_x)
        data_y = np.array(data_y)

        # 最后 look_back 天作为测试输入
        test = data_set[
            -self.look_back:,
            0
        ].reshape(1, 1, self.look_back)

        print('\n训练集 X shape:', data_x.shape)
        print('训练集 Y shape:', data_y.shape)

        return data_x, data_y, test


    # ==================================================
    # LSTM模型
    # ==================================================
    def rnn_model(self, train_x, train_y, epochs):

        model = Sequential()

        model.add(
            LSTM(
                64,
                input_shape=(1, self.look_back),
                return_sequences=True
            )
        )

        model.add(
            LSTM(
                32,
                return_sequences=False
            )
        )

        model.add(Dense(32))
        model.add(Dense(30))

        model.compile(
            loss='mean_squared_error',
            optimizer='adam'
        )

        model.summary()

        early_stopping = EarlyStopping(
            monitor='loss',
            patience=self.patience,
            restore_best_weights=True
        )

        history = model.fit(
            train_x,
            train_y,
            epochs=epochs,
            batch_size=self.batch_size,
            verbose=self.verbose,
            callbacks=[early_stopping]
        )

        return model


    # ==================================================
    # 预测
    # ==================================================
    def predict(self, model, data):

        prediction = model.predict(data)

        return prediction


    # ==================================================
    # 绘图
    # ==================================================
    def plot_show(self, predict):

        plt.figure(figsize=(18, 6))

        plt.plot(
            predict['report_date'],
            predict['total_purchase_amt'],
            label='purchase'
        )

        plt.plot(
            predict['report_date'],
            predict['total_redeem_amt'],
            label='redeem'
        )

        plt.legend()

        plt.title('Prediction Result')

        plt.grid(True)

        plt.show()


    # ==================================================
    # 主运行函数
    # ==================================================
    def run(self):

        # ======================================
        # purchase数据
        # ======================================
        purchase_train_x, purchase_train_y, \
        purchase_test, purchase_scaler = \
            self.access_data(self.purchase)

        # ======================================
        # redeem数据
        # ======================================
        redeem_train_x, redeem_train_y, \
        redeem_test, redeem_scaler = \
            self.access_data(self.redeem)

        # ======================================
        # 训练模型
        # ======================================
        print('\n开始训练 Purchase 模型...')
        purchase_model = self.rnn_model(
            purchase_train_x,
            purchase_train_y,
            self.epochs_purchase
        )

        print('\n开始训练 Redeem 模型...')
        redeem_model = self.rnn_model(
            redeem_train_x,
            redeem_train_y,
            self.epochs_redeem
        )

        # ======================================
        # 预测
        # ======================================
        purchase_predict = self.predict(
            purchase_model,
            purchase_test
        )

        redeem_predict = self.predict(
            redeem_model,
            redeem_test
        )

        # ======================================
        # 反归一化
        # ======================================
        purchase = purchase_scaler.inverse_transform(
            purchase_predict
        ).reshape(30)

        redeem = redeem_scaler.inverse_transform(
            redeem_predict
        ).reshape(30)

        # ======================================
        # 构造提交文件
        # ======================================
        predict_dates = pd.date_range(
            start='2014-09-01',
            periods=30
        )

        test_user = pd.DataFrame({

            'report_date': [
                int(x.strftime('%Y%m%d'))
                for x in predict_dates
            ],

            'total_purchase_amt': purchase.astype(int),

            'total_redeem_amt': redeem.astype(int)

        })

        # ======================================
        # 输出结果
        # ======================================
        print('\n预测结果：')
        print(test_user)

        # ======================================
        # 保存提交文件
        # ======================================
        if self.store_result:

            test_user.to_csv(
                '../outputs/submit_lstm.csv',
                index=False
            )

            print('\nsubmit_lstm.csv 已保存')

        # ======================================
        # 绘图
        # ======================================
        self.plot_show(test_user)


# ======================================================
# 主函数入口
# ======================================================
if __name__ == '__main__':

    initiation = RNNModel(

        look_back=40,

        epochs_purchase=150,

        epochs_redeem=230,

        batch_size=16,

        verbose=2,

        patience=50,

        store_result=True

    )

    initiation.run()
```
