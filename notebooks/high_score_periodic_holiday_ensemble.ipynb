# High-score Periodic and Holiday Model

This notebook builds a competition-oriented periodic-factor model with holiday handling and large-transaction segmentation. Model selection prioritizes an official-score proxy rather than MAPE alone.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path('..').resolve()
sys.path.insert(0, str(PROJECT_ROOT))

from src.high_score_periodic import (
    config_from_row,
    forecast_segmented,
    load_daily_segments,
    search_configs,
    write_submission,
)

RAW_ROOT = PROJECT_ROOT / 'data' / 'raw'
segments = load_daily_segments(RAW_ROOT)
{name: frame.shape for name, frame in segments.items()}
```

## Configuration Search

Use rolling 30-day validation windows before September 2014. Higher proxy score is better.

```python
VALIDATION_STARTS = [
    '2014-05-01',
    '2014-06-01',
    '2014-07-01',
    '2014-08-01',
]

search_result = search_configs(segments, VALIDATION_STARTS)
search_result.head(15)
```

## September Submission

Generate the official 30-row, headerless submission file.

```python
best_config = config_from_row(search_result.iloc[0])
print(best_config)

submission = forecast_segmented(segments, '2014-09-01', best_config, periods=30)
submission_path = PROJECT_ROOT / 'outputs' / 'submissions' / 'high_score_periodic' / 'tc_comp_predict_table.csv'
write_submission(submission, submission_path)
print(submission_path)
submission
```
